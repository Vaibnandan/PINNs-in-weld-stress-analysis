```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
data = pd.read_csv("p_chart_data.csv")

# Calculate the proportion of nonconforming items
data['Proportion'] = data['Nonconforming'] / data['Total']

# Calculate the overall proportion of nonconforming items
overall_proportion = data['Nonconforming'].sum() / data['Total'].sum()

# Create the P-chart
plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['Proportion'], marker='o', linestyle='-', color='b', label='Proportion')
plt.axhline(overall_proportion, color='r', linestyle='--', label='Overall Proportion')

# Customize the chart
plt.title('P-Chart')
plt.xlabel('Date')
plt.ylabel('Proportion')
plt.grid(True)
plt.legend()

# Show the chart
plt.show()
```
